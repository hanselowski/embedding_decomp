```python
def remove_duplicates(values):
    output = []
    seen = []
    for value in values:
        # If value has not been encountered yet,
        # ... add it to both list and set.
        if value not in seen:
            output.append(value)
            seen.append(value)
    return output

# prepare dataset
# path = r"/home/hanselowski/jupyther_projects/vecspace_decomp/vecspace_experiments/datasets/capitals_countries.csv"
path = r"/home/hanselowski/jupiter_projects/vecSpace_decomp/datasets/google_analogy.csv"

import csv
#with  as f:       
cr = csv.reader(open(path,"r", encoding='utf8'),delimiter=" ") # , is default
rows = list(cr)  # create a list of rows for instance

sets, names = [], []
set, cur_entry = [], "dummy"   
names.append(rows[0][1]+'_1')
names.append(rows[0][1]+'_2')
for row in rows:
    if not row[0] == cur_entry:     
        if ":" in row and len(set)>0 : 
            sets.append(set)
            names.append(row[1]+'_1')
            names.append(row[1]+'_2')
            # print(set)
            set = []
        elif len(row)>2:
            # print(row)
            set.append(row)
        # if ":" in row:
            # print(row)
    cur_entry = row[0]
sets.append(set) 
#print(set)  

# get category
cat_set = []
for s in sets:
    for i in range(2):
        cat_tmp = []
        for row in s:
            cat_tmp.append(row[i])
        cat_set.append(cat_tmp)


for cat in cat_set:
    print(cat)
    print(len(cat))

print(names)
print(len(names))
print(len(cat_set))

#########################################################
import sys
sys.path.append('..')
from utils.corpuscont import store_corpus, retrieve_corpus

# store corpus as a text file
store_corpus(cat_set, names, "../datasets/googe_cat_corpus.txt")

capitals = cat_set[2]
countries = cat_set[3]
print(capitals)
print(countries)
```
