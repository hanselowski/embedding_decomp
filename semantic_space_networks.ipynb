```python
# Load Google's pre-trained Word2Vec model.

model = gensim.models.KeyedVectors.load_word2vec_format('data/word2vec/GoogleNews-vectors-negative300.bin.gz', binary=True) 
print('... model loaded ...')
```

```python
########################################################################
# 3.2.1 Binary tree
# Example 4

print("")
print("binary tree")

sub_vec_1   = ['father','mother'] 
sub_vec_2 = ['brother','sister'] 

tree_1 = Tree(model, sub_vec_1, None)
tree_1.find_insts()
tree_1.branch_instances(0)
tree_1.branch_instances(1)

tree_2 = Tree(model, sub_vec_2, None)
tree_2.find_insts()
tree_2.branch_instances(0)
tree_2.branch_instances(1)

sub_vec = ['tree_1','tree_2'] 
human = Tree(model, sub_vec, [tree_1.root, tree_2.root])
human.find_insts()
human.branch_instances(0)
human.branch_instances(1)

```
